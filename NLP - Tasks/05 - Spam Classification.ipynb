```python
import pandas as pd
data = pd.read_csv("/content/drive/MyDrive/SpamData/SMSSpamCollection", sep='\t', names=["label", "message"])
data.head()
```

```python
#nltk libraries

import nltk
nltk.download('punkt')
nltk.download('stopwords')
nltk.download('wordnet')

from nltk.corpus import stopwords
from nltk.stem import PorterStemmer
```

```python
import re #regular expression
stemmer = PorterStemmer()
corpus = [] # corpus list

for i in range(0, len(data)):
    rev = re.sub('[^a-zA-Z]', ' ', data['message'][i])
    rev = rev.lower()
    rev = rev.split()
    rev = [stemmer.stem(word) for word in rev if not word in set(stopwords.words('english'))]
    rev = ' '.join(rev)
    corpus.append(rev)
```

```python
corpus[:10]
```

```python
#implement bag of words
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features=5000) #original feature count is 6296
X = cv.fit_transform(corpus).toarray()
```

```python
# X
```

```python
#Determine target data
y = pd.get_dummies(data['label']) # spam = 1 ham = 0
y = y.iloc[:,1].values #consider 1 column
y #dependent feature
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.20)
```

```python
#Use naive bayes - classification technique
from sklearn.naive_bayes import ComplementNB

spam_detect_model = ComplementNB().fit(X_train, y_train)
```

```python
pred = spam_detect_model.predict(X_test)
```

```python

from sklearn.metrics import confusion_matrix
confusion_m = confusion_matrix(y_test, pred)
```
